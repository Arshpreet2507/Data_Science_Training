CONVERTING THE DATA TYPES OF FOLLOWING :
1. REVIEWS TO NUMERIC
2. INSTALLS TO NUMERIC

```python
main_df[main_df['Reviews'].str.isnumeric() == False]
```

```python
main_df = main_df.drop(10472)
```

```python
main_df.loc[10470:10474]
```

```python
main_df['Reviews']=pd.to_numeric(main_df['Reviews'])
```

```python
main_df['Installs'] = (
    main_df['Installs']
    .str.replace(',', '')
    .str.replace('+', '', regex=False)
)
```

```python
main_df['Installs'] = pd.to_numeric(main_df['Installs'])
```

```python
main_df['Price'].unique()[:20]
```

```python
main_df['Price']= main_df['Price'].str.replace('$','',regex=False)
```

```python
main_df['Price']= pd.to_numeric(main_df['Price'])
```

```python
main_df['Last Updated']= pd.to_datetime(main_df['Last Updated'])
```

```python
main_df.info()
```

```python
main_df['Installs'] = pd.to_numeric(main_df['Installs'])
```

```python
main_df.info()
```

HANDLING MISSING VALUES 

```python
main_df.isnull().sum()
```

The dataset contained missing values in the Rating, Type, Current Ver, and Android Ver columns.

1. Missing values in Type were filled with "Free" because only one record was affected.
2. Rows with missing Current Ver and Android Ver values were removed because they represented less than 0.1% of the dataset.
3. Missing values in Rating were replaced with the median rating to preserve observations while minimizing the impact of missing data.

```python
main_df[main_df['Type'].isnull()]
```

```python
main_df['Type']=main_df['Type'].fillna('free')
```

```python
main_df.dropna(subset=['Current Ver'],inplace=True)
```

```python
main_df.dropna(subset=['Android Ver'], inplace =True)
```

```python
main_df['Rating'].median()
```

```python
main_df['Rating']= main_df['Rating'].fillna(
    main_df['Rating'].median()
)
```

```python
main_df.isnull().sum()
```

```python
main_df.duplicated().sum()
```

```python
main_df.drop_duplicates(inplace= True)
```

Question 1.Which category has the largest number of apps? 

```python
category_counts = main_df['Category'].value_counts()
category_counts.head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x=category_counts.head(10).values,
    y=category_counts.head(10).index
)
plt.title('Top 10 Categories by Number of Apps')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
```

Family, Game, and Tools categories contain the largest number of applications on the Google Play Store, indicating high competition in these categories.

Question 2. Which categories have the highest average ratings?

```python
avg_rating_df = (
    main_df.groupby('Category')['Rating']
    .mean()
    .sort_values(ascending=False)
)
avg_rating_df.head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x=avg_rating_df.head(10).values,
    y=avg_rating_df.head(10).index
)
plt.title('Top Categories by Average Rating')
```

Summary of Findings

The Google Play Store dataset was analyzed to understand app characteristics, user engagement, and category-wise trends. The dataset initially contained missing values and inconsistent data types, which were cleaned through preprocessing and data preparation techniques.

Several exploratory analyses and visualizations were performed on the cleaned dataset. The key findings from the analysis are:

The average app rating was approximately 4.2, indicating that most applications receive positive feedback from users.

The distribution of ratings showed that the majority of apps have ratings between 4.0 and 4.5.

The Family, Game, and Tools categories contained the highest number of applications, suggesting intense competition in these categories.

Free applications significantly outnumbered paid applications on the Google Play Store.

Applications with a higher number of reviews generally tended to have more installs, indicating a positive relationship between user engagement and app popularity.

The median app price was $0, confirming that most apps are available free of cost.

Install counts and review counts were highly skewed, with a small number of apps accounting for a large share of total downloads and reviews.




Conclusion
This exploratory data analysis provided valuable insights into the structure and characteristics of applications available on the Google Play Store. The analysis revealed that user preference strongly favors free applications and that a small number of highly popular apps dominate overall downloads and reviews. Category-wise analysis showed that certain categories attract more developers and users than others.
The visualizations helped identify patterns in ratings, installs, reviews, and category distributions, making it easier to understand app performance and user behavior. Overall, the dataset demonstrates the highly competitive nature of the mobile application market and highlights factors that may influence app popularity.
