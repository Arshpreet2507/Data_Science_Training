```python
import pandas as pd

df = pd.read_excel("samplesuperstore.csv.xlsx")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.drop('Row ID', axis=1, inplace=True)
```

```python
print(df.columns)
```

```python
df.isnull().sum()
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df[['Sales','Profit','Quantity','Discount']].describe()
```

```python
df.isnull().sum()
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
print("Rows and Columns:", df.shape)
```

```python
print(df.columns.tolist())
```

```python
df.describe()
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
```

```python
df[['Order Date','Month']].head()
```

```python
best_products = (
    df.groupby('Product Name')['Quantity']
      .sum()
      .sort_values(ascending=False)
)

print(best_products.head(10))
```

```python
import matplotlib.pyplot as plt

best_products.head(10).plot(kind='bar')
plt.title('Top 10 Best Selling Products')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
monthly_sales = (
    df.groupby('Month')['Sales']
      .sum()
)

print(monthly_sales)
```

```python
monthly_sales.plot(figsize=(10,5))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
top_customers = (
    df.groupby('Customer Name')['Sales']
      .sum()
      .sort_values(ascending=False)
)

print(top_customers.head(10))
```

```python
customer_orders = (
    df.groupby('Customer Name')['Order ID']
      .count()
      .sort_values(ascending=False)
)

print(customer_orders.head(10))
```

```python
total_revenue = df['Sales'].sum()

print("Total Revenue =", total_revenue)
```

```python
category_sales = (
    df.groupby('Category')['Sales']
      .sum()
      .sort_values(ascending=False)
)

print(category_sales)
```

```python
category_sales.plot(kind='bar')
plt.title('Revenue by Category')
plt.ylabel('Revenue')
plt.show()
```

```python
print("Total Profit =", df['Profit'].sum())
```

```python
profit_category = (
    df.groupby('Category')['Profit']
      .sum()
      .sort_values(ascending=False)
)

print(profit_category)
```

```python
region_sales = (
    df.groupby('Region')['Sales']
      .sum()
      .sort_values(ascending=False)
)

print(region_sales)
```

```python
region_profit = (
    df.groupby('Region')['Profit']
      .sum()
      .sort_values(ascending=False)
)

print(region_profit)
```

```python
discount_profit = (
    df.groupby('Discount')['Profit']
      .mean()
)

print(discount_profit)
```
